```python
# Generelle moduler og funksjonsbeskrivelser brukt i forelesningen
from numpy import sin, cos, pi, exp
import numpy as np
import scipy.signal as sig
import matplotlib.pyplot as plt
from Kildekode._12_IIRFilt import *


%matplotlib ipympl

def bodePlot(b, a, dt=None, f_max = 100, label=None): 
    if dt:
        w, mag, phase = sig.dlti(b, a, dt=dt).bode(w = 2*pi*np.logspace(-1, np.log10(1/dt/2), 501)*dt)
        plt.subplot(2,1,1)
        plt.semilogx(w/2/pi, mag, label=label)
        plt.plot([1/dt/2, 1/dt/2], [0, 0], 'C3:')[0].set_ydata([-1000, 1000])
        plt.ylabel(r'Magnitude (dB)')
        plt.legend()
        plt.grid(True)
        plt.subplot(2,1,2)
        plt.semilogx(w/2/pi, phase, label=label)
        plt.plot([1/dt/2, 1/dt/2], [0, 0], 'C3:')[0].set_ydata([-1000, 1000])
        plt.ylabel(r'Phase (degrees)')
        plt.xlabel(r'Frequency (Hz)')
        plt.legend()
        plt.grid(True)

    else:
        w, mag, phase = sig.lti(b, a).bode(w = 2*pi*np.logspace(-1, np.log10(f_max), 501))
        plt.subplot(2,1,1)
        plt.semilogx(w/2/pi, mag, label=label)
        plt.ylabel(r'Magnitude (dB)')
        plt.legend()
        plt.grid(True)
        plt.subplot(2,1,2)
        plt.semilogx(w/2/pi, phase, label=label)
        plt.ylabel(r'Phase (degrees)')
        plt.xlabel(r'Frequency (Hz)')
        plt.legend()
        plt.grid(True)
    #plt.title("Bode Plot")
```

# Noen analoge "standard-filtre"

    
|Filter:|Egenskaper:|
|---|---|
|**Butterworth**|"Maksimalt flat" respons i passbåndet <br> Konstant rolloff (20db/dec/filtord)|
|**Chebyshev type 1**|Ingen rippel i passbåndet <br>Rippel i stoppbåndet|
|**Chebyshev type 2**|Konstant rolloff i stoppbåndet <br> Rippel i passbåndet|
|**Elliptisk**|Minimalt transisjonsbånd <br>Rippel i passbånd og stoppbånd|
|**Bessel**| *Nesten* lineær faserespons <br>Lavest rolloff i stoppbånd|

# Lavpass filterdesign med Python

* `scipy.signal` inneholder funksjoner en rekke filtre, beskrevet [her](https://docs.scipy.org/doc/scipy/reference/signal.html#matlab-style-iir-filter-design)
* Kodecellen nedenfor viser et eksempel på 5 ulike framgangsmåter for å lage et 6. ordens lavpassfilter med knekkfrekvens $\hat{\omega}_c=0.4\pi$.

```python
from Kildekode._14_Z_transformasjon import displayFrequencyResponse

N = 6 # Filterorden
w_c = 0.4*pi

b1, a1 = sig.butter(N,              # Filterorden
                    Wn=w_c/pi,      # Knekkfrekvens 
                    btype='lowpass' # Filtertype
                   )

b2, a2 = sig.cheby1(N,              # Filterorden
                    rp=0.5,         # Maksimal tillatt rippel i passbånd (dB)
                    Wn=w_c/pi,      # Knekkfrekvens 
                    btype='lowpass' # Filtertype
                   )

b3, a3 = sig.cheby2(N,              # Filterorden
                    rs=30,          # Minimum stoppbåndsdemping (dB)
                    Wn=w_c/pi,      # Stoppbpånd Kantfrekvens
                    btype='lowpass' # Filtertype
                   )

b4, a4 = sig.ellip(N,                # Filterorden
                    rp=0.5,          # Maksimal tillatt rippel i passbånd (dB)
                    rs=30,           # Minimum stoppbåndsdemping
                    Wn=w_c/pi,       # Knekkfrekvens 
                    btype='lowpass'  # Filtertype
                   )
b5, a5 = sig.bessel(N,               # Filterorden
                    Wn=w_c/pi,       # Knekkfrekvens 
                    btype='lowpass'  # Filtertype
                   )

plt.close(7); plt.figure(7, figsize=(8,5))

displayFrequencyResponse(b1, a1, mag='lin', label='Butterworth')
displayFrequencyResponse(b2, a2, mag='lin', label='Chebychev type 1')
displayFrequencyResponse(b3, a3, mag='lin', label='Chebychev type 2')
displayFrequencyResponse(b4, a4, mag='lin', label='Elliptical')
displayFrequencyResponse(b5, a5, mag='lin', label='Bessel/Thompson')
```

# Inversfiltre/Rekonstruksjonsfiltre

* For et filter med transferfunksjon $H(z)$ kan det finnes et inversfilter $H^{-1}(z)= \frac{1}{H(z)}$.
* $H(z)$ og $ H^{-1}(z)$ i kaskade kansellerer hverandre.
* Forutsetter at både $H(z)$ og $H^{-1}(z)$ er stabile.
<img src="Figurer/14_Ztransform/Fig5_Reconstruction.png" style="width: 60%; margin-left: 150px" />
* Ofte referert til som rekonstruksjonsfilter.
* Grunnlag for konsept som ekkokansellering m.m.

# Komplementærfiltre

* ***Komplementærfilteret*** $H_{\text{comp}}(\hat{\omega})$ til et filter med styrkegradsrespons $\left| H(\hat{\omega}) \right|$ er et filter er styrkegradsresponsen er $\left|H_{\text{comp}}(\hat{\omega})\right| = 1 - \left| H(\hat{\omega}) \right|$
* For førsteordens IIR-filtre gjelder:
    $$H_{\text{comp}}(z) = 1 - H(z)$$
* Eksempel på bruk: **Sensor Fusion**.
    * Beregning av vinkelen til en drone basert på gyroskop og akselerometer. 
        * **Gyroskop** responderer bra til endringer i vinkel, men vil ofte ha en bias.
        * **Akselerometer** vil ha en god *gjennomsnittsverdi*, men plukker opp mye støy.
        * Ved å kombinere lavpassfiltrerte målinger fra akselerometer og høypassfiltrerte målinger fra gyroskopet, vil vi utnytte de beste egenskapene til begge sensorene.
    * For mer detaljert gjennomgang: [video om dronekontroll](https://www.youtube.com/watch?v=whSw42XddsU)

## Regneeksempel 2:
* Finn komplementærfilteret til lavpassfilteret $H(z) = \frac{0.1}{1-0.9\cdot z^{-1}}$

```python
# 
b1, a1 = [0.1], [1, -0.9] # Førsteordens lavpassfilter
b2, a2 = [0.9, -0.9], [1, -0.9] # Komplementærfilter

w, H1 = sig.freqz(b1, a1)
w, H2 = sig.freqz(b2, a2)

plt.close(5); plt.figure(5, figsize=(9,6))
plt.plot(w, np.abs(H1), label=r'$H(z) = \frac{0.1}{1-0.9\cdot z^{-1}}$')
plt.plot(w, np.abs(H2), label=r'$H(z) = \frac{0.9 - 0.9\cdot z^{-1}}{1-0.9\cdot z^{-1}}$')
plt.grid(True)
plt.xlim([0, pi])
plt.ylim(ymin=0)
plt.xticks(np.linspace(0, pi, 6), [r'$'+str(round(x,2))+r'\pi$' for x in np.linspace(0, 1, 6)] )
plt.xlabel(r"Normalisert Vinkelfrekvens $\hat{\omega}$")
plt.legend()
plt.show()
```

# Diskretisering:

# $$H(s) \rightarrow H(z)$$

# Eksempel på motivasjon 
* Implementasjon av en regulator vil svært ofte bruke en datamaskin som opererer med diskréte sampleverdier.
* Vi trenger å finne et *Digitalt Filter* som kan realisere en ønsket regulator med f.eks. transferfunksjon $G_c(s)$.
<img src="Figurer/16_DigitalKontroller/Fig1_sloyfe.png" style="width: 60%; margin-left: 200px" />

* Figuren har opphav fra et annet emne, og bruker noe annen notasjon. Fra og med neste slide bruker vi "vanlig" signalbehandlingsnotasjon.

# Hvordan diskretisere en transferfunksjon $H(s)$?
* Kort svar: [få python til å gjøre det](https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.cont2discrete.html)
    * Vi må allikevel ha en viss forståelse for hva python gjør, og hvilken metode vi vil bruke.
    * Det kan til tider være fordelaktig å ha et formeluttrykk som angir digitale filterkoeffisisenter uttrykt som funksjon av parametre definert i tidsdomenet (F.eks. $K_p$, $K_i$ og $K_d$- leddene i en PID-kontroller).


    
### De fleste fremgangsmåter faller i to hovedkategorier:
1. Finn en ***approksimasjon*** for laplace-operatoren $s$ i z-domenet, og bruk vanlig algebra til å finne den diskrete transferfunksjonen $H(z)$.
2. For et analogt filter $H(s)$ der en spesikk type inngangssignal $x(t)$ gir utgangssignalet $y(t)$, finn et digitalt filter $H(z)$ som produserer utgangssignalet $y(n\cdot T_s)$ når inngangssignalet er $x(n\cdot T_s)$.

# Kategori 1: substitusjon av $s$ med en numerisk tilnærming

* Enkleste metode: 
    * Bakoverdifferanse (Euler's backward method)
* Mest nøyaktige metode: 
    * Trapesintegrasjon (Bilineær Transformasjon / Tustin's metode)

# Bakoverdifferanse

* Løpende integral er sum av areal til rektangel som strekker seg bakover i tid fra sampeltidspunkt.

<img src="Figurer/16_DigitalKontroller/Fig13_Backdiff.png" style="width: 50%; margin-left: 15%" />

* Filteralgoritme for integrator med bruk av bakoverdifferanse:
$$y[n] = T_s \cdot x[n] + y[n-1]$$

* Utled transferfunksjon:
$$ \begin{align}
y[n] - y[n-1] &= T_s \cdot x[n] \\
Y(z) - z^{-1} \cdot Y(z) &= T_s \cdot X(z) \\
\frac{Y(z)}{X(z)} &= \frac{T_s}{1-z^{-1}} 
\end{align}$$

* Finn substitusjon for $s$: 
$$ \begin{align}
\frac{1}{s} &\approx \frac{T_s}{1-z^{-1}} \\
s &\approx \frac{1}{T_s} \cdot \frac{z-1}{z}
\end{align}$$


# Kommentarer til bakoverdifferanse

    <img src="Figurer/16_DigitalKontroller/Fig11_ZPbackdiff.png" align="right" width="30%" />

          <li>Diskretisering av filtre med kantfrekvenser relativt nære nyquist-frekvensen vil gi store avvik i egenskaper.</li>

          <li>Du ønsker å utlede et algebarisk uttrykk for digitale filterkoeffisienter $a_k$ og $b_k$ gitt kontinuerlige systemparametre, og andre metoder vil gi et unødvendig komplisert uttrykk</li>
        </ul>    
    </ul>  
    </div>
</div> 

## Eksempel på diskretisering med Python

```python
w_c = 10*20*pi # Knekkfrekvens for analogt filter 
b_c, a_c = sig.butter(3, w_c, analog=True) # Bruker python til å generere et analogt filter som utgangspunkt

b_c =  [w_c**3]
a_c = [1, 2*w_c, 2*w_c**2, w_c**3]
# Diskretisering med Fs=200 sampler per sekund
F_s = 2000    # Samplingsfrekvens
T_s = 1/F_s  # Sampleperiode
b_d, a_d, dt = sig.cont2discrete((b_c, a_c), 
                                  method='backward_diff', 
                                  dt=T_s
                                 )

# Sammenlign Bode-plot
plt.close(1);plt.figure(1, figsize=(9,6))
bodePlot(b_c, a_c , f_max=F_s,label='Kontinuerlig system')
bodePlot(b_d, a_d, dt=T_s, label='Bakoverdifferanse')

print(b_d[0])
print(a_d)

# Vis stegresponsen:
plt.close(2);plt.figure(2, figsize=(9,3))
t, s_t = sig.lti(b_c, a_c).step(T=np.linspace(0, 0.02, 501))
plt.plot(t, s_t)
t_n, s_n = sig.dlti(b_d, a_d, dt=T_s).step(n=F_s*1)
plt.plot(t_n, np.squeeze(s_n), drawstyle="steps-post")
plt.xlim([0, 0.02])
plt.grid(True)
plt.xlabel('Tid $t$ (sekund)')
plt.tight_layout()
```

* Legg merke til hvordan avvik mellom kontinuerlig og diskret system blir mer merkbart desto større filterets knekkfrekvens er.

# Bilineær Transformasjon (*basert på trapesintegrasjon*)

* Anbefalt metode for diskretisering av en regulator
* Løpende integral er areal under kurve med lineær interpolasjon mellom sampler.

<img src="Figurer/16_DigitalKontroller/Fig14_Trapezoidal.png" style="width: 70%; margin-left: 15%" />

* Filteralgoritme for integrator med bruk av bakoverdifferanse:
$$y[n] = T_s \cdot \left(\frac{x[n] + x[n-1]}{2}\right) + y[n-1]$$

* Utled transferfunksjon:
$$ \begin{align}
y[n] - y[n-1] &= \frac{T_s}{2} \cdot x[n]+\frac{T_s}{2} \cdot x[n-1] \\
Y(z) - z^{-1} \cdot Y(z) &= \frac{T_s}{2} \cdot X(z) + \frac{T_s}{2} \cdot z^{-1} \cdot X(z) \\
\frac{Y(z)}{X(z)} &= \frac{T_s}{2}\cdot\frac{1 + \cdot z^{-1}}{1-z^{-1}} 
\end{align}$$

* Finn substitusjon for $s$: 
$$ \begin{align}
\frac{1}{s} &\approx \frac{T_s}{2}\cdot\frac{1 +  z^{-1}}{1-z^{-1}}  \\
s &\approx \frac{2}{T_s} \cdot \frac{z-1}{z+1}
\end{align}$$

# Frekvens-"warping" 

* Forholdet mellom et punkt på frekvensaksen $s = j\omega$, og et punkt på enhetssirkelen $z=e^{j\hat{\omega}}$ er ***ikke*** lineær. Forholdet mellom de beskrives med formelen
$$ \omega = \frac{2}{T_s} \cdot \tan \left( \frac{\hat{\omega}}{2} \right)$$

<img src="Figurer/16_DigitalKontroller/Fig15_BilinearFreq.png" style="width: 70%; margin-left: 10%" />

* Dersom en spesifikk kantfrekvens skal ivaretas ved diskretisering, må man korrigere for dette.

# Pre-warping

* Metode for å tilpasse frekvens-fordreinginen slik at én spesifikk kantfrekvens beholder sin plass.
* Ikke innebygd i `scipy.signal`, men kan enkelt utføres ved å regne ut en tilsvarende fordreining av *sampletiden* $T_s$ som brukes til transformasjonen.
    1. Finn den normaliserte vinkelfrekvensen $\hat{\omega}_c$ til kantfrekvensen $\omega_c$ som skal beholdes.
    2. Regn ut hvilken vinkelfrekvens $\omega_c^*$ som *faktisk* vil mappes til $\hat{\omega}_c$ med bruk av formelen i forrige slide.
    3. Regn ut fordreid sampletid $T_s^* = \frac{\omega_c^*}{\omega_c}$, og benytt denne i substitusjonen av $s$.

## Regneeksempel 1:

* Et høypassfilter med kantfrekvens $f_c = 70$Hz, skal diskretiseres med et system som bruker samplingsfrekvens $f_s = 200$Hz. 
    * Hvilken kontinuerlige kantfrekvens vil *egentlig* gi kantfrekvensen $\hat{\omega}_c = 2\pi \cdot \frac{70}{200}$ etter diskretisering?
    * Hva kan vi sette sampleperioden til for å ivareta denne kantfrekvensen?

## Eksempel på diskretisering med Python

```python
w_c = 50*2*pi # Knekkfrekvens for filter
b_c, a_c = sig.ellip(4, 1, 30, w_c, analog=True, btype='lowpass')
F_s = 200
T_s = 1/F_s

# Utfør pre-warping ved å finne en justert samplingsfrekvens
w_prewarp = 2/T_s*np.tan(w_c*T_s/2)
warp_factor = w_prewarp/(w_c)
dt_warped = T_s*warp_factor

b_d, a_d, dt = sig.cont2discrete((b_c, a_c), 
                                        method='bilinear', 
                                        dt=dt_warped
                                       )
# Sammenlign Bode-plot
plt.close(3);plt.figure(3, figsize=(9,6))
bodePlot(b_c, a_c , f_max=200,label='Kontinuerlig system')
bodePlot(b_d, a_d, dt=T_s, label='Bilineær Transformasjon')

# Vis stegresponsen:
plt.close(4);plt.figure(4, figsize=(9,3))
t, s_t = sig.lti(b_c, a_c).step(T=np.linspace(0, 0.3, 501))
plt.plot(t, s_t)
t_n, s_n = sig.dlti(b_d, a_d, dt=T_s).step()
plt.plot(t_n, np.squeeze(s_n), drawstyle="steps-post")
plt.xlim([0, 0.3])
plt.grid(True)
plt.xlabel('Tid $t$ (sekund)')
plt.tight_layout()
```

# Oppsummering Bilineær Transformasjon

* Utføres med substitusjonen $s \approx \frac{2}{T_s}\cdot \frac{z-1}{z+1}$
* Gir mest presis ivaretagelse av frekvensegenskapene til et filter.
* Pre-warping bør benyttes dersom filteret har kantfrekvens $\omega_c > \pi \cdot \frac{f_s}{10}$
* Ikke like godt egnet til system med fler enn 1 kantfrekvens.
* Stabilt kontinuerlig system $\leftrightarrow$ stabilt diskrét system.

# Kategori 2: approksimasjon av kjent utgangssignal

* **Mest relevante metode**:
    * Step Invariant Method (a.k.a. Zero Order Hold approksimasjon)
        * *Henter navnet fra Zero Order Hold D/A omformingsmetoden, men dette er to **forskjellige** tema.*
        * *Anbefalt metode for diskretisering av en prosess som skal reguleres*
    * Gitt et analogt filter med stegrespons $s(t)$, så finner denne metoden det digitale filteret der stegresponsen $s[n] = s(t)\big|_{t = n\cdot T_s}$
        * Baserer seg på avbildning av *poler* fra s-domenet til z-domenet.
        
<img src="Figurer/16_DigitalKontroller/Fig16_StepResponse.png" style="width: 70%; margin-left: 10%" />

* Vi begynner med å se på førsteordens system

# Poler i S-planet og stegrespons
* Én pol $s=p$ i s-planet representerer et førseordens filter / system.
    * Transferfunksjon: $H(s) = \frac{p}{s-p}$
    * Stegrespons: $s(t) = \left(1-e^{p\cdot t}\right) \cdot u(t)$
    
    
<img src="Figurer/16_DigitalKontroller/Fig2_Spole.png" style="width: 70%" />

# Poler i Z-planet og stegrespons
* Én pol $z=\hat{p}$ i Z-planet representerer et førsteordens IIR-filter.
    * Transferfunksjon: $H(z) = \frac{1-\hat{p}}{z-\hat{p}}=\frac{(1-\hat{p})z^{-1}}{1-\hat{p}\cdot z^{-1}}$
        * Eksempel: $\hat{p} = 0.9 \rightarrow H(z) = \frac{0.1}{z-0.9}$<br><br>
    * Filteralgoritme: $y[n] = (1-\hat{p})\cdot x[n-1] + \hat{p}\cdot y[n-1]$
    * Stegrespons: $s[n] = \left(1-\hat{p}\ ^n \right)\cdot u[n]$<br>

    
    
<img src="Figurer/16_DigitalKontroller/Fig3_Zpole.png" style="width: 70%" />

# Førsteordenssystem

* Sampling av impulsresponsen $s(t) = \left(1-e^{p\cdot t}\right)\cdot u(t)$:
\begin{align}
s[n] &= s(n\cdot T_s) \\
&= \left(1-e^{p\cdot n \cdot T_s}\right) \cdot u[n] \\
&= \left(1-\left(e^{p\cdot T_s}\right)^n\right) \cdot u[n]
\end{align}
* Utregning av transferfunksjon $H(z)$:
\begin{align}
s[n] &= \left(1-\left(e^{p\cdot T_s}\right)^n\right) \cdot u[n]\\
&= \left(1-\hat{p}\ ^n\right) \cdot u[n]\\
&\downarrow \\
H(z) &= \frac{1-\hat{p}}{z-\hat{p}}= \frac{1-e^{p \cdot T_s}}{z-e^{p \cdot T_s}}\\
\end{align}


* Her er det med vilje ikke et nullpunkt i $z=0$. Dette for å sikre at første sample i stegresponsen er lik null ($s[0] = 0$)

## Regneeksempel 2:

* Finn filterkoeffisientene til et digitalt filter med samme stegrespons som det tidskontinuerlige systemet med transferfunksjon $H(s) = \frac{10}{s +10 }$ når samplingsfrekvensen er $F_s = 20$.

```python
b_c, a_c = [10], [1, 10]
F_s = 20
T_s = 1/F_s

b_d, a_d, dt = sig.cont2discrete((b_c, a_c), 
                                        method='zoh', 
                                        dt=T_s
                                       )
print("b:",b_d, "\na:", a_d)

# Vis stegresponsen:
plt.close(5);plt.figure(5, figsize=(9,6))
t, s_t = sig.lti(b_c, a_c).step()
plt.plot(t, s_t)
t_n, s_n = sig.dlti(b_d, a_d, dt=T_s).step()
plt.plot(t_n, np.squeeze(s_n), drawstyle="steps-post")
plt.xlim([0, 1])
plt.grid(True)
plt.xlabel('Tid $t$ (sekund)')
```

# Hva med nullpunkt?
* Et nullpunkt $z=z_k$ i Z-planet representerer et førsteordens FIR filter.
    * Transferfunksjon $H(z) = 1-z_k\cdot z^{-1}$ 
    * Impulsrespons $h[n] = \delta[n] - z_k\cdot \delta[n-1]$.
    * Ikke realiserbar med fysisk system!
    * Det finnes ingen åpenbar parallell mellom nullpunktsplassering i $s$-planet og nullpunktsplassering i $z$-planet

* Hvordan diskretiseres da høyereordens system?

# Steginvarians-metoden (step invariant method):

### Gitt en kjent transferfunksjon $H(s)$ for et kontinuerlig system:

1. Finn den laplacetransformerte sprangresponsen ved å multiplisere med $\frac{1}{s}$.
2. Finn et *funksjonsuttrykk* for sprangresponsen $s(t)$ ved invers laplacetransformasjon:
$$s(t) = \mathcal{L}^{-1}\left(\frac{1}{s}\cdot H(s) \right) $$
    * Her brukes typisk delbrøkoppspalting, polynomdivisjon og annen algebraisk manipulasjon for å oppnå en sum av førsteordens system med kjente transformasjonspar.
3. Diskretisér sprangresponsen $s(t)$ med hjelp av substitusjonen $s[n] = s(t)\big|_{t=n\cdot T_s}$.
4. Regn ut den Z-transformerte av sprangresponsen $s[n]$.
$$S(z) = \mathcal{Z}\left(s[n]\right)$$
5. *Differensier* sprangresponsens Z-transformerte for å finne impulsresponsens z-transformerte, altså transferfunksjonen $H(z)$:
$$H(z) = S(z)\cdot \left(1-z^{-1}\right)$$

#### Oppsummert:
$$H(z) = \mathcal{Z}\left(\mathcal{L}^{-1}\left(\frac{1}{s}\cdot H(s) \right)\bigg|_{t=n\cdot T_s}\right)\cdot \left(1-z^{-1}\right)$$

# Eksempel på diskretisering med python

```python


w_c = 10*2*pi # Knekkfrekvens for filter
b_c, a_c = sig.ellip(4, 1, 30, w_c, analog=True)
F_s = 200
T_s = 1/F_s

b_d, a_d, dt = sig.cont2discrete((b_c, a_c), 
                                        method='zoh', 
                                        dt=T_s
                                       )

plt.close(6);plt.figure(6, figsize=(9,6))
bodePlot(b_c, a_c , f_max=200,label='Kontinuerlig system')
bodePlot(b_d, a_d, dt=T_s, label='Step Invariant Method')

# Vis stegresponsen:
plt.close(7);plt.figure(7, figsize=(9,3))
t, s_t = sig.lti(b_c, a_c).step()
plt.plot(t, s_t)
t_n, s_n = sig.dlti(b_d, a_d, dt=T_s).step()
plt.plot(t_n, np.squeeze(s_n), drawstyle="steps-post")
plt.xlim([0, 0.5])
plt.grid(True)
plt.xlabel('Tid $t$ (sekund)')
plt.tight_layout()
```
