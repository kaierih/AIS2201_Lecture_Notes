# Digitale LTI-system

* ***Alle*** kausale LTI-system kan beskrives med en generell differanselikning:

### $$ \sum_{k=0}^{N}a_k\cdot  y[n-k] = \sum_{k=0}^{M} b_k \cdot x[n-k]$$

* $a$ og $b$ er tallrekker som inneholder de ulike koeffisientene til systemet/filteret.
\begin{align}
a_k &\in \{a_0, a_1,\ldots , a_N\}\\
b_k &\in \{b_0, b_1, \ldots, b_M\}
\end{align}

# Impulsrespons og FIR Filtre

* **Impulsrespons $h[n]$:** Utgangssignalet til et digitalt LTI-system når inngangssignalet er en impuls $\delta[n]$.

<img src="Figurer/10_Digitale Filtre Intro/Fig10_hn.png" style="width: 40%; margin-left: 300px%" />

* **FIR filtre:** **F**inite **I**mpulse **R**esponse

* Generell differanseligning for FIR filtre:
### $$ y[n] = \sum_{k=0}^{M} b_k \cdot x[n-k]$$

* Impulsresponsen til et FIR filter er et signal som gjengir filterkoeffisientene $b_k$ etter tur.
\begin{align}
h[n] &= \sum_{k=0}^{M} b_k \cdot \delta[n-k]\\
     &= b_0 \cdot \delta[n] + b_1\cdot \delta[n-1] + \ldots + b_M\cdot \delta[n-M]
\end{align}

<img src="Figurer/10_Digitale Filtre Intro/Fig11_FIR.png" style="width: 80%; margin-left: 100px%" />

# Diskrét Konvolusjon

* Utgangssignalet til et LTI-system utregnes ved å utføre ***konvolusjon*** av inngangssignalet $x[n]$ og impulsresponsen $h[n]$.
    * Konvolusjon har en egen matematisk operator: $*$.

\begin{align}
y[n] &= \sum_{k=0}^{\infty} x[k]\cdot h[n-k]\\
&= x[n]*h[n]
\end{align}

```python
from Kildekode._10_Konvolusjon import convolutionDemo
n = np.arange(60); xn = np.cos(pi/6*n) # x[n] er en Sinussekvens
hn = (-0.75)**np.arange(10)               # h[n] = 0.75^n, 0 <= n < 10.
convolutionDemo(xn, hn, fig_num=2,figsize=(10,7));    
```

# Konvolusjon egenskaper

1. Kommutativ 
\begin{align}
x[n]*y[n] &= y[n]*x[n] \\
\sum_{k=-\infty}^{\infty} x[k]\cdot y[n-k] &= \sum_{k=-\infty}^{\infty} y[k]\cdot x[n-k]
\end{align}


2. Assossiativ
$$(x[n]*y[n])*v[n] = x[n]*(y[n]*v[n])$$


3. Distributiv *(linearitet)*
$$(x[n] + y[n])*v[n] = x[n]*v[n] + y[n]*v[n]$$


4. Konvolusjon og fouriertransformasjon
$$ x[n]*y[n] \leftrightarrow X\left(\hat{\omega} \right)\cdot Y\left(\hat{\omega} \right)$$

# Frekvensrespons

* ***Frekvensresponsen*** $H\left(\hat{\omega}\right)$ til et digitalt filter er det samme som frekvensinnholdet til impulsresponsen $h[n]$.
    * Filtrering av et signal $x[n]$ er det samme som å multiplisere signalets frekvensinnhold $X\left(\hat{\omega}\right)$ med frekvensresponsen $H\left(\hat{\omega}\right)$.

* $H\left(\hat{\omega}\right)$ er en *kompleks funksjon* som forteller hvordan filteret vil påvirke enhver ren sinussekvens $x[n] = A\cdot \cos(\hat{\omega}_0\cdot n + \phi)$.
<img src="Figurer/11_Frekvensrespons/Fig3_LTI_SInusoid.png" style="width: 80%; margin-left: 100px" />

* Når vi analyserer filtre er det gunstig å kunne studere skaleringen og forskyvningen hver for seg. Disse to "komponentene" er kjent som amplituderesponsen og faseresponsen til et filter.
    * **Styrkegradsrespons** $\left|H\left(\hat{\omega}\right)\right|$
        * Forteller hvor mye en sinussekvens forsterkes/dempes, gitt digital frekvens $\hat{\omega}$.
    * **Faserespons** $\angle H\left(\hat{\omega}\right)$
        * Forteller hvor mange radianer sinussekvensen ut av filteret er forskjøvet i forhold til inngangssignalet, gitt digital frekvens $\hat{\omega}$.

# IIR Filtre i tidsdomenet

* **IIR** - **I**nfinite **I**mpulse **R**esponse
* Fellesbetegnelse for filter som kalkulerer utgangsveriden $y[n]$ basert på en lineær kombinasjon av verdiene til inngangssignalet ved foregående sampletidspunkt $x[n-k], \ \ k\in \{0, 1, 2, \ldots\}$ ***og*** tidligere kalkulerte verdier til utgangssignalet $y[n-k], \ \ k \in \{1, 2, \ldots\}$.
    * Har en tydelig parallell til analoge filtre.
    * Har to sett med koeffisienter: $a_k$ og $b_k$.


* **Differanseligning:**
$$\sum_{k=0}^{N} a_k \cdot y[n-k] = \sum_{k=0}^{M} b_k \cdot x[n-k]$$
* **Filteralgoritme:**
$$y[n] = \frac{1}{a_0}\cdot \left(\sum_{k=0}^{M} b_k \cdot x[n-k] - \sum_{k=1}^{N}a_k\cdot  y[n-k] \right)$$

* Vi bruker funksjonen `scipy.signal.lfilter` til å utføre IIR-filtrering i Python.

# Z-transformasjon

* "Laplacetransformasjon for digitale filtre"

#### Notasjon:
$$\textbf{Z}(h[n]) = H(z)$$

#### Formel definisjon:
\begin{align}
H(z) &= \sum_{n=0}^{\infty} h[n]\cdot z^{-n}\\
& \text{der}\\
z &= r\cdot e^{j\hat{\omega}}
\end{align}

* Verdien til funksjonen $H(z)$ i punktet $z$ forteller hvordan filteret vil respondere til et inngangssignal $x[n] = z^n$.


#### Nøkkelegenskap *tidsforskyving*:
* Z-transformasjon lar oss transfomrere *differanseligninger* til *polynomuttrykk*:
$$\textbf{Z}(x[n-D]) = z^{-D}\cdot X(z)$$

# Transferfunksjon / overføringsfunksjon

* Z-transformen av impulsresponen $h[n]$ til et LTI-system kalles systemets *transferfunksjon*
* For ethvert kausalt LTI-system med filterkoeffisienter **a** og **b**, vil transferfunksjonen være:
\begin{align}
H(z) = \frac{Y(z)}{X(z)} &= \frac{\sum_{k=0}^{M} b_k \cdot z^{-k}}{\sum_{k=0}^{N} a_k \cdot z^{-k}}\\
&\text{eller}\\
&= \frac{b_0 + b_1\cdot z^{-1} + b_2 \cdot z^{-2} + \ldots + b_M \cdot z^{-M}}{a_0 + a_1\cdot z^{-1} + a_2 \cdot z^{-2} + \ldots + a_N \cdot z^{-N}}
\end{align}
* Teller og nevner er polynomer
* Frekvensresponsen *leses av* ved å følge transferfunksjonen langs enhetssirkelen: 
$$ H(\hat{\omega}) = H(z)\big|_{z = e^{j\hat{\omega}}}$$

# Design av IIR filtre
* Designprosess: Design et filter i $s$-domenet, og så *diskretisér*.
    * *Anvendt i blant annet reguleringsteknikk for implementasjon av kontrollere på digitale enheter.*
### Framgangsmåter:
1. Substitusjon $s \rightarrow f(z)$ i filterets transferfunksjon.
    * Backward Euler: $s \rightarrow f_s \cdot \frac{z-1}{z}$
        * Fin å bruke dersom $f_s >> f_c$, og når *enkle* regneuttrykk for filterkoeffisisenter er en fordel.
    * Trapesmetoden/Bilineær Transformasjon/Tustin's metode: $s \rightarrow 2f_s \cdot \frac{z-1}{z+1}$
        * *NB: Husk pre-warping dersom $f_s< 5\cdot f_c$ (tommelfingerregel)*
        * Anbefalt metode for diskretisering av reguleringssystem
2. Sampling av utgangssignal.
    * Step invariant method/zero order hold:
        $$u[n] = u(t)\big|_{t = n\cdot T_s}$$
        * *NB: Introduserer ekstra forsinkelse* 
        * Anbefalt metode for simulering av tidskontinuerlige system.
