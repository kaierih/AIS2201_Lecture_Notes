# Potential areas for improvement:

### Frequency resolution
* *Errors up to $5Hz$ make it unsuitable for many practical applications*

### Noise reduction
* *If the system is listening for specific notes in a noisy invironment, the noise may cause a number of false detections*

### Ability to handle non-sinusoidal signals
* *Most real-world, highly periodic signals (such as sound) are not pure sines, but composed of a fundamental frequency and a number of harmnonics.*
* *As we saw in exercise 1, the strongest frequency component is not necessarily the fundamental frequency*

# Proposed system parameters:



* The final system should produce a fundamental frequency estimate $\hat{f}_0$ of a signal within the frequency range $25\text{Hz} \leq f \leq 4200\text{Hz}$.
    * It is not a requirement to be able to handle signals which fall outside this category as a separate "case".

* The estimate "refresh interval" must be at most $0.2$ seconds, although it is permissable for samples taken more than $0.2$ seconds ago to influence the estimate.

* The algorithm must be able to adapt to changes in fundamental frequency within $0.5$ seconds, with samples more than $0.5$ seconds old having negligible influence (e.g. $\leq 10\%$).

* You are free to choose any sampling frequency you may wish, as long as it is achievable on the STM32 hardware.
